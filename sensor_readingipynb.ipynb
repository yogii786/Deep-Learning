```
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv('/content/drive/My Drive/sensor_readings_24.csv', header= None)
```

```
df.shape
```

```
df.head()
```

```
df['motion'] = df[24]
df = df.drop(24, axis= 1)
df.head()
```

```
x = df.iloc[:,:-1]
y = df.iloc[:,-1]

```

```
Motion = pd.get_dummies(y, drop_first = True)
```

```
y = pd.concat([y,Motion], axis = 1)
y = y.drop(['motion'],axis = 1)
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```
x_train.shape
```

```
set(y_train)
```

```
model = Sequential()
```

```
model.add(Dense(32, input_dim=24))

model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4,activation='softmax'))
```

```
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
model.fit(x_train,y_train,epochs=10,batch_size=200)
```
